```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Load the csv
us_arrests_df = pd.read_csv("UsArrests.csv")
# Display the dataframe
us_arrests_df
```

Show the most populated cities

```python
city_population = us_arrests_df[["UrbanPop", "City"]].sort_values(by=["UrbanPop"], ascending=False)
city_population.head(10)
```

```python
plt.bar(city_population["City"], city_population["UrbanPop"])
plt.xticks(rotation=90)
plt.xlabel("City")
plt.ylabel("Urban Population")
plt.title("Urban Population by City")
plt.show()
```

The top 5 most populated cities are: California, New Jersey, Rhode Island, New York and Massachusetts

Explore the murders by city

```python
city_murders = us_arrests_df[["Murder", "City"]].sort_values(by=["Murder"], ascending=False)
city_murders.head(10)
```

```python
plt.bar(city_murders["City"],city_murders["Murder"])
plt.xticks(rotation=90)
plt.xlabel("City")
plt.ylabel("Murder Rate")
plt.title("The cities with the highest murder incidents")
plt.show()
```

The top 3 cities with the highest murder incidents are: Georgia, Mississipi and Florida

Explore the cities with the highest assault incidents

```python
city_assaults = us_arrests_df[["Assault", "City"]].sort_values(by=["Assault"], ascending = False)
city_assaults.head(10)
```

```python
# Create a horizontal bar plot
fig, ax = plt.subplots(figsize=(10, 8))
ax.barh(city_assaults["City"], city_assaults["Assault"])

# Set plot title and axis labels
ax.set_title("Assaults by City")
ax.set_xlabel("Assault")
ax.set_ylabel("City")

plt.show()
```

The top 3 cities with the highest assault incidents are: North Carolina, Florida and Maryland

Explore rape by city

```python
rape_capital = us_arrests_df[["Rape", "City"]].sort_values(by=["Rape"], ascending=False)
rape_capital.head(10)
```

```python
# Create a horizontal bar plot using Seaborn
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(10,8))
sns.barplot(x="Rape", y="City", data=rape_capital, color="salmon")

# Set plot title and axis labels
ax.set_title("Rape Incidents by City")
ax.set_xlabel("Rape")
ax.set_ylabel("City")

plt.show()
```

The top 3 cities city with the highest rape incidents are: Nevada, Alaska and California
